# Select Features

```python
# Set features. This will also be used as your x values.
selected_features = df[['koi_fpflag_nt', 'koi_fpflag_ss', 'koi_fpflag_co',
       'koi_fpflag_ec', 'koi_period', 'koi_period_err1', 'koi_period_err2',
       'koi_time0bk', 'koi_time0bk_err1', 'koi_time0bk_err2', 'koi_impact',
       'koi_impact_err1', 'koi_impact_err2', 'koi_duration',
       'koi_duration_err1', 'koi_duration_err2', 'koi_depth', 'koi_depth_err1',
       'koi_depth_err2', 'koi_prad', 'koi_prad_err1', 'koi_prad_err2',
       'koi_teq', 'koi_insol', 'koi_insol_err1', 'koi_insol_err2',
       'koi_model_snr', 'koi_tce_plnt_num', 'koi_steff', 'koi_steff_err1',
       'koi_steff_err2', 'koi_slogg', 'koi_slogg_err1', 'koi_slogg_err2',
       'koi_srad', 'koi_srad_err1', 'koi_srad_err2', 'ra', 'dec',
       'koi_kepmag']]

selected_features.head(10)
```

# Create A Train Test Split 

Use koi_disposition for the y values

```python
# create x and y variables. y is koi disposition and x is selected features
y = df["koi_disposition"]
X = selected_features
```

```python
# add train test split, random subset of the data. Where 70% will be used to train the model to find an equation that fits.
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=7, stratify=y)
```

```python
X_train.head()
```

# Pre-processing

Scale the data using the MinMaxScaler and perform some feature selection

```python
# dependencies
# scaling of the data bring every data to the same scale, use the minmax scaler
from sklearn.preprocessing import MinMaxScaler
```

```python
# create the scalers
# Scale your data
X_minmax = MinMaxScaler().fit(X_train)

X_train_minmax = X_minmax.transform(X_train)
X_test_minmax = X_minmax.transform(X_test)
```

# Train the Model

```python
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier()
rf.fit(X_train_minmax, y_train)
```

```python
# passing the scaling data, everything should match for format/scaling/etc.
print(f"Training Data Score: {rf.score(X_train_minmax, y_train)}")
print(f"Testing Data Score: {rf.score(X_test_minmax, y_test)}")
```

```python
sorted(zip(rf.feature_importances_, selected_features), reverse=True)
```

# Hyperparameter Tuning

Use GridSearchCV to tune the model's parameters

```python
# Create the GridSearchCV model, use logistic regression cause it's a classifier, decision tree narrow down check hyper parameter tuning for best tuning parameters
# keep running the sample through different modesl again and again, changing the parameters when passed


from sklearn.model_selection import GridSearchCV

param_grid = {'n_estimators': [250, 300, 350],
             'max_depth': [125, 150, 175]}
grid = GridSearchCV(rf, param_grid, verbose=3)
```

```python
# Train the model with GridSearch

grid.fit(X_train_minmax, y_train)
```

```python
print(grid.best_params_)
print(grid.best_score_)
```
